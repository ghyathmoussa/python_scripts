```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dizi1 = np.linspace(0,10,20)
dizi2 = np.linspace(0,10,20)

dizi2 = dizi2 * 10*3


figure = plt.figure()
axes = figure.add_axes([0.2,0.2,0.9,0.9]) #left start point,right start point
                                          #left end point ,right end point
axes.plot(dizi2,dizi1)
```

```python
figure2 = plt.figure()
axes = figure2.add_axes([0.1,0.1,0.7,0.7])
axes2 = figure2.add_axes([0.1,0.6,0.3,0.3])

# first figure
axes.plot(dizi1,dizi2,'g')
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('Main Graph')

#second figure
axes2.plot(dizi2,dizi1,'r')
axes2.set_xlabel('x label')
axes2.set_ylabel('y label')
axes2.set_title('Small graph')
```

```python
new_fig = plt.figure()
new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])

new_axes.plot(dizi1,dizi2,label = 'dizi1')
new_axes.legend()
```
