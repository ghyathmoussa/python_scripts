```
# import packages
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
session = SparkSession.builder.appName('Ghyath Test').getOrCreate()
```

```
rowDF = session.read.format('kafka').option('kafka.bootstrap.servers','137.184.213.30:9092') \
                                    .option('subscribe','ghyath_test') \
                                    .option('starttingOffsets','earliest').load()
rowDF.show()
```

```
# rowDf look like pandas Data Frame
# handle data from value column
# create schema for data
schema = StructType() \
            .add('model',StringType()) \
            .add('year',StringType()) \
            .add('price',StringType()) \
            .add('transmission',StringType()) \
            .add('mileage',StringType()) \
            .add('fuelType',StringType()) \
            .add('tax',StringType()) \
            .add('mpg',StringType()) \
            .add('engineSize',StringType())
```

```
# get the data from kafka to spark sql
kafkaDf = rowDF.select(from_json(col('value').cast('string'),schema).alias('data'),'timestamp').select('data.*','timestamp')
kafkaDf.show()
# group data by name
```
